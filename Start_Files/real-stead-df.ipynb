```python
df_no_dupl = df.drop_duplicates(subset=None, keep='first', ignore_index = True)
df_no_dupl[df_no_dupl.duplicated()]
```

```python
df_no_dupl['Locality'].unique()
for vill in df_no_dupl['Locality'].unique():
    for vill2 in df_no_dupl['Locality'].unique():
        if vill[0:4] == vill2[0:4]:
            if vill != vill2:
                print(f"{vill} and {vill2}")
```

```python
all_cols = df_no_dupl.columns
colls_to_keep = ['Locality', 'Type of property', 'Subtype of property', 'Price',
       'Number of bedrooms', 'Livable surface', 'Kitchen equipment', 'Balcony',
       'State of the property', 'Furnished', 'Number of facades', 
       'Terrace', 'Surface terrace', 'Total land surface', 'Garden', 'Surface garden',
       'Fireplace', 'Swimming pool']
colls_to_drop = []
for item in all_cols:
    if item not in colls_to_keep:
        colls_to_drop.append(item)
df_18_base = df_no_dupl.drop(colls_to_drop, axis=1, inplace=False)
df_18_base
```

```python
#df['Kitchen equipment'].value_counts()
df_18_base = df_18_base.dropna(subset = ['Price', 'Number of bedrooms', 'Furnished'])
df_18_base['Swimming pool'] = df_18_base['Swimming pool'].fillna(0)
df_18_base.isna().sum()
```
