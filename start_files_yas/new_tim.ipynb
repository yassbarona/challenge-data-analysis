```python
import pandas as pd
pd.set_option('display.max_seq_items', None)

read_none = ["None"]
df = pd.read_csv('Final_result_wo_duplicates.csv', na_values = read_none)
df
```

```python
all_cols = df.columns
colls_to_keep = ['Immoweb ID', 'Post code', 'Property type', 'property sub-type', 'Price (sr only)',
       'Bedrooms', 'Living area', 'Kitchen type',
       'Building condition', 'Furnished', 'Number of frontages', 
       'Terrace', 'Terrace surface', 'Surface of the plot', 'Garden', 'Garden surface',
       'How many fireplaces?', 'Swimming pool']
colls_to_drop = []
for item in all_cols:
    if item not in colls_to_keep:
        colls_to_drop.append(item)
df = df.drop(colls_to_drop, axis=1, inplace=False)
df
```

```python
df = df.dropna(subset = ['Price (sr only)'])
df['Swimming pool'] = df['Swimming pool'].fillna('No')
df.isna().sum()
```

```python
postcode_dict = {}

def add_postcodes(name:str, begin:int, end:int):
    for i in range(begin, end +1):
        postcode_dict[i] = name

add_postcodes("Bruxelles", 1000, 1299)
add_postcodes("Brabant Wallon", 1300, 1499)
add_postcodes("Vlaams Brabant", 1500, 1999)
add_postcodes("Antwerpen", 2000, 2999)
add_postcodes("Vlaams Brabant", 3000, 3499)
add_postcodes("Limburg", 3500, 3999)
add_postcodes("Liège", 4000, 4999)
add_postcodes("Namur", 5000, 5999)
add_postcodes("Hainaut", 6000, 6599)
add_postcodes("Luxembourg", 6600, 6999)
add_postcodes("Hainaut", 7000, 7999)
add_postcodes("West-Vlaanderen", 8000, 8999)
add_postcodes("Oost-Vlaanderen", 9000, 9999)


df['Region'] = df['Post code'].map(postcode_dict)
df
```

```python
import plotly.express as px
from pandasgui import show

key_cols = [val for val in ['property sub-type', str('Post code'), None, None] if val is not None]
if key_cols != []:
    if 'mean' is not None:
        df = df.groupby(key_cols).agg('mean').reset_index()
    else:
        df = df.sort_values(key_cols)
fig = px.bar(data_frame=df, y='Post code', x='Price (sr only)', color=
    'property sub-type', facet_row=None, facet_col=None, )
show(fig)

```
